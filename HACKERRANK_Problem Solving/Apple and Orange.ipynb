```python
s = int(input('Enter a no. (s)'))
t = int(input('Enter a no. (t)'))
a = int(input('Enter a no. (a)'))
b = int(input('Enter a no. (b)'))
apples = int(input('Enter a no. (m)'))
oranges = int(input('Enter a no. (n)'))
#print(a)
#print(b)
#print(arr)
#print(brr)

main =[]
for i in range(s, t+1):
    main.append(i)
#print(main)


accum = [] 

for i in range(0, apples): 
    ele = int(input()) 
  
    accum.append(ele) # adding the element 
      
#print(accum)

accum2 = [] 

for i in range(0, oranges): 
    elb = int(input()) 
    
    accum2.append(elb) # adding the element 
      
#print(accum2)



arr = []
brr = []
ab = 0
bc = 0
acb = 0
acd = 0

for i in accum:
    ab = a + i
    arr.append(ab)
print(arr)

for j in accum2:
    bc = b + j
    #print(bc)
    brr.append(bc)
print(brr)



for c in arr: 
    for d in main:
        if c == d:
            acb+=1
print(acb)


for o in brr: 
    for p in main:
        if o == p:
            acd+=1
print(acd)

           
```
